Data Type Validation & Format Standardization

```python
read_xsl.info()
```

```python
print(read_xsl["Product"].head())
```

```python
read_xsl["Product"] = read_xsl["Product"].str.strip()
```

```python
read_xsl["Product"] = read_xsl["Product"].str.title()
```

```python
text_columns = [
    "Product",
    "ShippingAddress",
    "PaymentMethod",
    "OrderStatus",
    "ReferralSource"
]

for col in text_columns:
    read_xsl[col] = read_xsl[col].astype(str).str.strip().str.title()
```

Date Standardization

```python
print(read_xsl["Date"].head())
```

```python
read_xsl["Date"] = pd.to_datetime(read_xsl["Date"])
```

```python
read_xsl["Date"] = read_xsl["Date"].dt.strftime("%Y-%m-%d")
```
